```python
df1.loc[df1['StudyTimeWeekly'] < 5.5, 'StudyTimeWeekly'] = 0
df1.loc[(df1.StudyTimeWeekly >=5.5) &  (df1.StudyTimeWeekly<11), 'StudyTimeWeekly'] = 1
df1.loc[(df1.StudyTimeWeekly >=11) &  (df1.StudyTimeWeekly<16.5), 'StudyTimeWeekly'] = 2
df1.loc[df1.StudyTimeWeekly>=16.5, 'StudyTimeWeekly'] = 3
```

```python
df1
```

```python
df1.StudyTimeWeekly=df1.StudyTimeWeekly.astype(dtype="int")
```

```python
x=df1.drop(["GradeClass","Ethnicity","ParentalEducation","GPA","Sports","Volunteering"],axis="columns")
y=df1.GradeClass
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(x,y,test_size=0.2)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=200)
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
predictions = model.predict(X_test)
```

```python
# Calculate Mean Squared Error 
# -- If youâ€™re using a classification Random Forest (RandomForestClassifier), MSE is not typically used â€” instead, youâ€™d use accuracy, precision, recall, or log loss.
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, predictions)
rmse = np.sqrt(mse)  # Root Mean Squared Error (optional)

print(f"Mean Squared Error: {mse:.4f}")
print(f"Root Mean Squared Error: {rmse:.4f}")
```

```python
output = pd.DataFrame({'StudyTimeWeekly': X_test.StudyTimeWeekly, 'Age': X_test.Age, 'GradeClass': predictions})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```
